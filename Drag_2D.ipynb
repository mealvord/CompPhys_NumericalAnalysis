```python
#Incorprating air resistance into projectile motion

#import stuff
import numpy as np
import matplotlib.pyplot as plt

#set parameters 
t_max = 5 #s, this will be the end of the simulation 
t_step = 0.001 #s

#define knowns
m = 10 #kg
g = np.array([0,-9.8]) #N/kg or m/s/s
rho = 1.225 #kg/m^3, density of fluid (air)
C = 0.47 #drag coefficient (unit less) sphere
r = .65 #m radius of sphere
A = np.pi*r**2 #m^2, cross sectional area

#define starting values (initial conditions)
pos = np.array([0,0]) #m
v = np.array([30,30]) #m/s
t = 0 #s

#creating lists to track values
x_list = [pos[0]] #hold my x position values
y_list = [pos[1]] #hold my y position values
vx_list = [v[0]] #hold my x velocity values
vy_list = [v[1]] #hold my x velocity values
t_list = [t] #hold my t values



while t <= t_max: 
    vix = vx_list[-1] #calls the last value of vx list
    viy = vy_list[-1] #calls the last value of vy list
    vi = np.array([vix,viy])
    vmag = np.linalg.norm(vi)
    vhat = vi / vmag
    
    posi = np.array([x_list[-1],y_list[-1]])
    
    Fnet = (0.5*rho*(vmag**2)*C*A)*(-vhat) + m*g
    a = Fnet / m
    
    vf = vi + a*t_step
    posf = posi + vf*t_step
    t = t + t_step
    
    v = np.vstack(vf) #attaches the most recent velocity array to the array
    
    
    y_list.append(posf[1])
    t_list.append(t)
    
plt.scatter(t_list,y_list)
plt.show()
plt.scatter(y_list,t_list)
plt.show()
```

```python
#Projectile with drag

#imports
import numpy as np
import matplotlib.pyplot as plt

#set simulation parameters
t_max = 5 #s, will be the end time of the simulation 
t_step = 0.001 #s

#define knowns
m = 10 #kg
g = np.array([0,-9.8]) #N/kg
rho = 1.225 #kg/m^3
C = 0.47 #unitless
r = 0.25 #m radius
A = np.pi*r**2 #m^2

#starting positions
pos = np.array([0,0]) #m
v = np.array([3,3]) #m/s
#print(v)
t = 0 #s

#creating lists to track the position and time over time
x_list = [pos[0]]
y_list = [pos[1]]
vx_list = [v[0]]
vy_list = [v[1]]
t_list = [t]
#a_list = [a]


while y_list[-1]>=0:
    vix = vx_list[-1]
    viy = vy_list[-1]
    vi = np.array([vix,viy])
    vmag = np.linalg.norm(vi)
    vhat = vi/vmag
    
    posi = np.array([x_list[-1],y_list[-1]])
    
    Fnet = m*g + 0.5*rho*(vmag**2)*C*A*-vhat
    a = Fnet / m   
    
    vf = vi + a*t_step
    posf = posi + vf*t_step
    t = t + t_step
    
    v = np.vstack(vf)
    pos = np.vstack(posf)
        
    x_list.append(posf[0])
    y_list.append(posf[1])
    vx_list.append(vf[0])
    vy_list.append(vf[1])
    t_list.append(t)
    #a_list.append(a)
    
   
plt.scatter(t_list,x_list)
plt.show()
plt.scatter(t_list,y_list)
plt.show()
plt.scatter(t_list,vx_list)
plt.show()
plt.scatter(t_list,vy_list)
plt.show()
plt.scatter(x_list,y_list)
plt.xlabel('horizontal position (m)')
plt.ylabel('vertical position (m)')
plt.title('Trajectory with Drag')
plt.show()
```
