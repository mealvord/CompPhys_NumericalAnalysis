```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([-3,-2,-1,0,1,2,3])
y = np.array([-.0335, -0.0205, -0.008,0, 0.008,0.0205,0.0335])

a,b = np.polyfit(x,y,deg=1)

plt.scatter(x,y)
plt.plot(x,a*x+b)
plt.xlabel('wave number')
plt.ylabel('peak location (m)')
plt.text(-2, 0.02, f'y = {b:.5f} + {a:.5f}m * m', size=14)
plt.show()
```

```python
d = 0.3e-3 #m
L = 2.95 #m

wavelength = a*d/L
print(wavelength, ' m')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4.539,5.076,5.310,6.383])*10**14
y = np.array([1.45025,1.5815,1.947,2.04])
y_J = y*1.6e-19

a,b = np.polyfit(x,y_J,deg=1)

plt.scatter(x,y_J)
plt.plot(x,a*x+b)
plt.xlabel('frequency (Hz)')
plt.ylabel('Energy (J)')
plt.title('Planck’s Constant Measured from LED Observations')
plt.text(5.5e14, 2.5e-19, f'E = {a:.3e}Js* f + {b:.3e}J', size=14)
plt.show()
print(f'E = {a:.3e}Js* f + {b:.3e}J')
```

```python
print('experimental slope = ', a, 'Js')
expected = 6.62607015e-34
print('expected slope = ',expected,' Js')
percenterror = np.abs(a - expected)/expected * 100
print('percent error = ',percenterror,'%')
```
